```python
movies['genres']=movies['genres'].apply(json.loads)
for index,i in zip(movies.index,movies['genres']):
    list1=[]
    for j in range(len(i)):
        list1.append((i[j]['name']))
    movies.loc[index,'genres']=str(list1)

movies['keywords']=movies['keywords'].apply(json.loads)
for index,i in zip(movies.index,movies['keywords']):
    list1=[]
    for j in range(len(i)):
        list1.append((i[j]['name']))
    movies.loc[index,'keywords']=str(list1)

# movies['production_companies']=movies['production_companies'].apply(json.loads)
# for index,i in zip(movies.index,movies['production_companies']):
#     list1=[]
#     for j in range(len(i)):
#         list1.append((i[j]['id']))
#     movies.loc[index,'production_companies']=str(list1)
      
movies['production_countries']=movies['production_countries'].apply(json.loads)
for index,i in zip(movies.index,movies['production_countries']):
    list1=[]
    for j in range(len(i)):
        list1.append((i[j]['iso_3166_1']))
    movies.loc[index,'production_countries']=str(list1)
```

```python
movies['production_companies']=movies['production_companies'].apply(json.loads)
def production_company(x):
    list1 = []
    for i in x:
        if(i['id'] not in list1):
            list1.append(i['id'])
    return list1
movies['production_companies']=movies['production_companies'].apply(production_company)
```

```python
mov['cast']=mov['cast'].apply(json.loads)
def main_cast(x):
    cnt = 0
    list1 = []
    #只保留主演演员
    for i in x:
        if len(list1)<=10:
            if(i['id'] not in list1):
                list1.append(i['id'])
        else:
            return list1
    return list1
mov['main_cast']=mov['cast'].apply(main_cast)
```

```python
mov['crew']=mov['crew'].apply(json.loads)
#Crew部分除了director都按department分类
# director
def director(x):
    list1 = []
    for i in x:
        if i['job'] == 'Director':
            if(i['id'] not in list1):
                list1.append(i['id'])
    return list1
mov['director']=mov['crew'].apply(director)
```

```python
def editing(x):
    list1 = []
    for i in x:
        if i['department'] == 'Editing':
            if(i['id'] not in list1):
                list1.append(i['id'])
    return list1
mov['editing']=mov['crew'].apply(editing) 
```

```python
def sound(x):
    list1 = []
    for i in x:
        if i['department'] == 'Sound':
            if(i['id'] not in list1):
                list1.append(i['id'])
    return list1
mov['sound']=mov['crew'].apply(sound) 
```

```python
def art(x):
    list1 = []
    for i in x:
        if i['department'] == 'Art':
            if(i['id'] not in list1):
                list1.append(i['id'])
    return list1

def production(x):
    list1 = []
    for i in x:
        if i['department'] == 'Production':
            if(i['id'] not in list1):
                list1.append(i['id'])
    return list1

def visualeffects(x):
    list1 = []
    for i in x:
        if i['department'] == 'Visual Effects':
            if(i['id'] not in list1):
                list1.append(i['id'])
    return list1

def costume(x):
    list1 = []
    for i in x:
        if i['department'] == 'Costume & Make-Up':
            if(i['id'] not in list1):
                list1.append(i['id'])
    return list1

def camera(x):
    list1 = []
    for i in x:
        if i['department'] == 'Camera':
            if(i['id'] not in list1):
                list1.append(i['id'])
    return list1
def crew(x):
    list1 = []
    for i in x:
        if i['department'] == 'Crew':
            if(i['id'] not in list1):
                list1.append(i['id'])
    return list1

def lighting(x):
    list1 = []
    for i in x:
        if i['department'] == 'Lighting':
            if(i['id'] not in list1):
                list1.append(i['id'])
    return list1

def writing(x):
    list1 = []
    for i in x:
        if i['department'] == 'Writing':
            if(i['id'] not in list1):
                list1.append(i['id'])
    return list1

def directing(x):
    list1 = []
    for i in x:
        if i['department'] == 'Directing':
            if(i['id'] not in list1):
                list1.append(i['id'])
    return list1
```

```python
mov['art']=mov['crew'].apply(art) 
mov['production']=mov['crew'].apply(production) 
mov['visual_effects']=mov['crew'].apply(visualeffects) 
mov['costume']=mov['crew'].apply(costume) 
mov['camera']=mov['crew'].apply(camera)  
mov['lighting']=mov['crew'].apply(lighting) 
mov['writing']=mov['crew'].apply(writing) 
mov['crew_member']=mov['crew'].apply(crew)
mov['directing']=mov['crew'].apply(directing)
```

```python
# 这个算法太慢&维度太高
def uniqueList(col):
    List = []
    for index, row in movies.iterrows():
        col_all = row[col]
        for cell in col_all:
            if cell not in List:
                List.append(cell)
    if List[-1] == '':
        return List[0:-1]
    else:
        return List

def binary(unique_list, cell_list):
    binaryList=[]
    for one in unique_list:
        if one in cell_list:
            binaryList.append(1)
        else:
            binaryList.append(0)
    return binaryList
```

```python
genresList = uniqueList('genres')
```

```python
castList = uniqueList('main_cast')
```

```python
movies['genres_bin'] = movies['genres'].apply(lambda x: binary(genresList,x))
```
